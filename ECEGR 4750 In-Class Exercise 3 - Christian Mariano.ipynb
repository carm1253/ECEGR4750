Regression Line w/ NOX as the response variable

```python

"""
regression example
Chirag Shah, University of Washington
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

from google.colab import files
files.upload()

df = pd.read_csv("Boston.csv", index_col = 0)
print("correlation coefficient = ", np.corrcoef(df.age, df.nox) [0,1])

y = df.nox
X = df.age
X = sm.add_constant(X)

lr_model = sm.OLS(y,X).fit()

print(lr_model.summary())

X_prime = np.linspace(X.age.min(), X.age.max(), 100)

X_prime = sm.add_constant(X_prime)

y_hat = lr_model.predict(X_prime)

plt.scatter(df.age, df.nox)

plt.plot(X_prime[:,1], y_hat, "red", alpha = 0.9)
plt.xlabel("Proportion of Owner-Occupied Units Built Prior to 1940")
plt.ylabel("Nitric Oxides Concentration (Parts Per 10 million)")
```

5. For "nox", the independent variable should be "age" because more older occupied units could mean more nitric oxide concentrations since older units made be found in more urbanized or industrial areas that have higher NOX levels.The graph shows that age and nox are positively correlated with a correlation coefficient of about 0.7 (which shows a strong correlation).

```python

"""
regression example
Chirag Shah, University of Washington
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

from google.colab import files
files.upload()

df = pd.read_csv("Boston.csv", index_col = 0)
print("correlation coefficient = ", np.corrcoef(df.rm, df.medv) [0,1])

y = df.medv
X = df.rm
X = sm.add_constant(X)

lr_model = sm.OLS(y,X).fit()

print(lr_model.summary())

X_prime = np.linspace(X.rm.min(), X.rm.max(), 100)

X_prime = sm.add_constant(X_prime)

y_hat = lr_model.predict(X_prime)

plt.scatter(df.rm, df.medv)

plt.plot(X_prime[:,1], y_hat, "red", alpha = 0.9)
plt.xlabel("Average Number of Rooms per Dwelling")
plt.ylabel("Median value of owner-occupied homes in $1000's")
```
